# **Sales Json**

**Load the data in json format**

```python
a = json.loads(js)
a
```

**Add all the sales Products_Id**

```python
Products = [Product_Id1,Product_Id2,Product_Id3,Product_Id4,Product_Id5]
```

**Add the Quantity details**

```python
Quantity = [Quantity1,Quantity2,Quantity3,Quantity4,Quantity5]
```

**Now convert into json format**

```python
sales = {}
for prod in range(len(Products)):
    sales[Products[prod]]={"Category" :a[Products[prod]]["Category"],
          "Type" :a[Products[prod]]["Type"],
          "Brand" :a[Products[prod]]["Brand"],
          "Model" :a[Products[prod]]["Model"],
          "Color" :a[Products[prod]]["Color"],
          "Quantity" :Quantity[prod],
          "Price" :a[Products[prod]]["Price"]*Quantity[prod]}
```

```python
sales
```

**Now dump the sale's data into json file**

```python
a = json.dumps(sales)
```
